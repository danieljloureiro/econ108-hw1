## Part I. HICP data from SDW

```python
# import pandas (and other libraries you may want to use)
import pandas as pd
```

```python
# import the data files you downloaded from SDW
infl0511 = pd.read_csv('data/ICP.M.U2.N.051100.4.ANR.csv', skiprows=6, header=None, usecols = [0,1])
infl0512 = pd.read_csv('data/ICP.M.U2.N.051200.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0513 = pd.read_csv('data/ICP.M.U2.N.051300.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0520 = pd.read_csv('data/ICP.M.U2.N.052000.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0531 = pd.read_csv('data/ICP.M.U2.N.053100.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0532 = pd.read_csv('data/ICP.M.U2.N.053200.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0533 = pd.read_csv('data/ICP.M.U2.N.053300.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0540 = pd.read_csv('data/ICP.M.U2.N.054000.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0551 = pd.read_csv('data/ICP.M.U2.N.055100.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0552 = pd.read_csv('data/ICP.M.U2.N.055200.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0561 = pd.read_csv('data/ICP.M.U2.N.056100.4.ANR.csv', skiprows=6, header=None, usecols = [1])
infl0562 = pd.read_csv('data/ICP.M.U2.N.056200.4.ANR.csv', skiprows=6, header=None, usecols = [1])
```

```python
# Print as a list the codes of the subindices you downloaded.
# For example, if you download items from the __FOOD AND NON-ALCOHOLIC BEVERAGES__ category, 
# list all codes type `01XX00` you were able to download. 
list_codes = ['051100','051200','051300','052000','053100','053200','053300','054000','055100','055200','056100','056200']
print(list_codes)
```

```python
# create a single dataframe combining data for all HICP subindices you downloaded
infl05 = pd.concat([infl0511, infl0512, infl0513, infl0520, infl0531, infl0532, infl0533, infl0540, infl0551, infl0552, infl0561, infl0562], axis=1)
infl05
```

```python
# make other changes to the dataframe - set datetime index, rename the columns (up to you, but shortning very names may be a good idea)
infl05.rename(columns={0:'Date'}, inplace=True)
infl05['Date'] = pd.to_datetime(infl05['Date'], format='%Y%b')
infl05.set_index('Date', inplace=True)
infl05.columns = list_codes
infl05
```

```python
# print the type of the index of the dataframe
print(type(infl05.index))
```

```python
# print the data types of the columns in the dataframe
print(infl05.dtypes)
```

```python
# save as a csv file under the name of the item group
infl05.to_csv(f'data/HICP_FURNISHINGS_HOUSEHOLD_EQUIPMENT_AND_ROUTINE_HOUSE_MAINTENANCE.csv', index=True)
```

```python
# compute and print the means and standard deviations for each series for the full period
out = infl05.agg(['mean','std'])
print(out)
```

```python
# compute and print the means and standard deviations for each calendar month for each series
out_month = infl05.groupby([infl05.index.month]).agg(['mean', 'std'])
print(out_month)
```

```python
# redo the above for the period between January 2017 and December 2021
infl05_1721 = infl05.loc['2017-01':'2021-12']
out1721 = infl05_1721.agg(['mean','std'])
print(out1721)
```

```python
out_month1721 = infl05_1721.groupby([infl05_1721.index.month]).agg(['mean', 'std'])
print(out_month1721)
```

## Part II. GDP data from EUROSTAT

```python
# import the data files you downloaded from EUROSTAT
GDP = pd.read_csv('data/namq_10_gdp.csv', usecols = [7,8])
GDP
```

```python
# create a dataframe with a datetime index matching the timing of the series you downloaded
GDP.columns = ['Date', 'GDP']
GDP['Date'] = pd.to_datetime(GDP['Date'])
GDP.set_index('Date', inplace=True)
GDP.sort_index(ascending=False, inplace=True)
GDP
```

```python
# save as a csv file under the name "Real-GDP-EA.csv"
GDP.to_csv(f'data/Real-GDP-EA.csv', index=True)
```

## PART III Combine GDP and inflation data

```python
# create a new datafame containing the GDP data and the inflation indices from the first task.
df1 = pd.read_csv('data/HICP_FURNISHINGS_HOUSEHOLD_EQUIPMENT_AND_ROUTINE_HOUSE_MAINTENANCE.csv', index_col=0)
df2 = pd.read_csv('data/Real-GDP-EA.csv', index_col=0)
dfmerge = pd.concat([df1, df2], axis=1)
dfmerge
```
